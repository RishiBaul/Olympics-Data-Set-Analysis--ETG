# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
df=pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
temp=df[df['Medal']=='Gold']
temp.groupby(['Sport'],as_index=False).size().to_frame('Gold Medals').sort_values(by='Gold Medals',ascending=False).head().plot(kind='bar',figsize=(7,7))
```

#OR

```python
lst=df[df['Medal']=='Gold']
coun=[]
for sport in lst['Sport'].unique():
  coun.append([sport,len(lst[lst['Sport']==sport])])

pd.DataFrame(coun,columns=['Sport','Gold Medals']).sort_values(by="Gold Medals", ascending=False).head().plot(x='Sport',y='Gold Medals',kind='bar', figsize=(7,7) )
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport',as_index=False).size().to_frame('Medals').sort_values(by='Medals',ascending=False).head().plot(kind='bar',figsize=(7,7))
```

#OR

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (7,7))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for ath in df['Athlete'].unique():
  data.append([ath,len(df[df['Athlete']==ath])])
pd.DataFrame(data,columns=['Athlete','Medals']).sort_values(by='Medals', ascending=False).head().plot(x='Athlete',y='Medals',kind='bar',figsize=(7,7))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
temp=df[df['Medal']=='Gold']
data=[]
for ath in temp['Athlete'].unique():
  data.append([ath,len(temp[temp['Athlete']==ath])])
pd.DataFrame(data,columns=['Athlete','Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x='Athlete',y='Gold Medals',kind='bar',figsize=(7,7))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
temp=df[df['Medal']=='Gold']
res=temp[temp['Country']=='IND']
min(res['Year'])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby('Event',as_index=False).size().to_frame('Players').sort_values(by='Players',ascending=False).head().plot(kind='bar',figsize=(7,7))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
temp=df[df['Gender']=='Women']
res=temp[temp['Medal']=='Gold'].groupby(['Sport'],as_index=False).size().to_frame('Female Gold Medalists').sort_values(by='Female Gold Medalists', ascending=False)
res.sort_values(by='Female Gold Medalists',ascending=False).head().plot(kind='bar', figsize=(7,7))
```
